```python
import pandas as pd
d=pd.read_csv("main.csv");
d
```

```python
nandf=d.loc[d.isna().any(axis=1)]
nandf
```

```python
df1=d.dropna()
```

```python
splits = list(df1.groupby("product_description"))
len(splits)
```

```python
splits[0][0]
```

```python
di={};
for i in range(0,len(splits)):
  di[splits[i][0]]=sum((splits[i][1]['price']*splits[i][1]['sales_quantity']))/sum(splits[i][1]['sales_quantity']);
di['apple']
```

```python
p_new=[0]*nandf.shape[0];
nandf['price_new']=p_new;
for i in range(0,len(splits)):
  nandf.loc[nandf["product_description"] == splits[i][0], "price_new"] = di[splits[i][0]];
nandf
```
